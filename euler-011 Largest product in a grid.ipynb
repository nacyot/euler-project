```python
def right(row, col)
  [[row, col], [row, col + 1], [row, col + 2], [row, col + 3]]
end

def down(row, col)
  [[row, col], [row + 1, col], [row + 2 , col], [row + 3, col]]
end

def digonal(row, col)
  [[row, col], [row + 1, col + 1], [row + 2 , col + 2 ], [row + 3, col +3]]
end

def rdigonal(row, col)
  result = [[row, col], [row + 1, col - 1], [row + 2 , col - 2 ], [row + 3, col -3]]
  result.map do |index|
    index[1] = index[1] >= 0 ? index[1] : nil
  end
  return result
end

right(1, 1)
```
